# Procedural Programming Examples

# Sequential Execution
Code runs line by line, top to bottom

```python
x = 10
y = 20

result = x + y
print(f"The sum is: {result}")

x = x * 2
new_result = x + y
print(f"The product is: {new_result}")
```

# Mutable State
Variable change over time

```python
counter = 0
total = 0

counter += 1
total += 10
print(f"After first update: counter = {counter}, total = {total}")

counter += 1
total += 20
print(f"After second update: counter = {counter}, total = {total}")

average = total / counter
print(f"Average: {average}")
```

# Procedures and Functions


```python
def print_greeting(name):
    print(f"Hello, {name}!")
    print("Welcome to the program.")


def calculate_area(length, width):
    area = length * width
    return area


print_greeting("Frank")
result = calculate_area(5, 10)
print(f"Area: {result}")
```

# Loops

```python
for i in range(5):
    print(f"Iteration {i}")

# While loop
count = 0
while count < 5:
    print(f"Count: {count}")
    count += 1

# Iterate over list
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)
```

# Conditional


```python
def check_grade(score):
    if score >= 90:
        grade = 'A'
    elif score >= 80:
        grade = 'B'
    elif score >= 70:
        grade = 'C'
    elif score >= 60:
        grade = 'D'
    else:
        grade = 'F'
    return grade


scores = [95, 82, 74, 58]
for score in scores:
    grade = check_grade(score)
    print(f"Score {score}: Grade {grade}")
```

# Modifying Collections

```python
numbers = [1, 2, 3]
print(f"Original: {numbers}")

numbers.append(4)
numbers.append(5)
print(f"After append: {numbers}")

numbers.remove(2)
print(f"After remove: {numbers}")

numbers[0] = 10
print(f"After modification: {numbers}")

# Dictionary manipulation
person = {'name': 'John', 'age': 30}
person['city'] = 'New York'
person['age'] = 31
print(f"Person: {person}")
```

# Side Effects


```python
total_processed = 0

def process_item(item):
    global total_processed
    total_processed += 1
    print(f"Processing: {item}")
    return item * 2

items = [10, 20, 30]
results = []

for item in items:
    result = process_item(item)
    results.append(result)

print(f"Results: {results}")
print(f"Total processed: {total_processed}")
```
