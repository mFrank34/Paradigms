# Ex.type file handling
create a file, write  list of names into it, then read and print the names


```python
import os

path = "fileHandling.txt"
names = {"michael", "theo", "john"}

os.remove(path)

# creating a file
file = open(path, "x")

# write names into file
with open(path, "w") as f:
    for name in names:
        f.write(name + "\n")

with open(path) as f:
    print(f.read())
```

# Ex.type 8: Problem-solving
1. Given a list of numbers, find the largest prime number.
2. Given a list of numbers, find the maximum value.
3. Write a Python program to find the sum of all prime numbers less than a given
number.
4. Calculate and print the sum of all prime numbers less than the given limit

```python
numbers = list(range(1, 50))
primeNums = []

def isPrime(num):
    for index in range(2, num):
        if num % index == 0:
            return False
    return True

def sumLessPrime(limit, primeNums):
    total = 0
    for prime in primeNums:
        if prime < limit:
            total += prime
    return total

for number in numbers:
    if number < 2:
        continue

    if isPrime(number):
        primeNums.append(number)

print(f"Larges prime number: {primeNums[len(primeNums) - 1]}")

print(f"Larges number in the list: {numbers[len(numbers) - 1]}")

print(f"Sum of all prime number up to 25: {sumLessPrime(25, primeNums)}")
```
