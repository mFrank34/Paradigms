## Ex.type 3 : Conditional Statement

Check if a given number is even or odd and print the result.
Implement conditional statements for decision-making: Write a Python program
that checks if a given number is positive, negative, or zero.

```python
# --- conditional statements ---

def isOdd(num):
    for index in range(2, num):
        if num % index == 0:
            return False
    return True


# --- setup ---
number = 6

if isOdd(number):
    print(f"{number} is odd")
else:
    print(f"{number} is even")
```

## Ex.type 7: File handling
Create a file, write a list of names into it, and then read and print the names.

```python
import os


def write(file, content):
    with open(file, "a") as f:
        for item in content:
            f.write(item + "\n")


def read(file):
    with open(file, "r") as f:
        for line in f:
            print(line)


path = "practical2.txt"
names = ["michael", "theo", "john", "Tony", "matthew"]

write(path, names)
read(path)

os.remove(path)
```

## Ex.type 8: Problem-solving
1. Given a list of numbers, find the largest prime number.
2. Given a list of number, find the maximum value
3. Write a Python program to find the sum of all prime numbers less than a given
number.
4. Calculate and print the sum of all prime numbers less than the given limit

```python
def is_prime(number):
    if number <= 1:
        return False
    else:
        for i in range(2, int(number ** 0.5) + 1):
            if number % i == 0:
                return False
    return True


def sum_less(limit, primeNums):
    total = 0
    for prime in primeNums:
        if prime <= limit:
            total += prime
    return total


def cal(numbers):
    prime_Nums = []
    for i in numbers:
        if not is_prime(i):
            prime_Nums.append(i)
    return prime_Nums


# --- setup ---
numbers = list(range(1, 25))
primeNums = cal(numbers)

# --- output ---
print(primeNums)

print(f"The largest prime number is: {primeNums[-1]}")
print(f"The maximum Value in numbers is: {numbers[-1]}")

less_than = 5
print(f"The sum of all prime numbers less then {less_than} is {sum_less(less_than, primeNums)}")
```
