# Exercise 2: Guessing Game with Limits
1. Generate a random number between 1 and 100.
2. Get the user's guess and validate it to ensure it's within the valid range (1-100).
3. Check if the guess is correct.
4. Provide hints (higher or lower) if the guess is incorrect.

```python
# Guessing game
import random as rmd


def game(input, target):
    if input == target:
        print("you win the game!")
        return False
    elif input > target:
        print(f"{input}: lower")
        return True
    elif input < target:
        print(f"{input}: higher")
        return True


def main():
    target = rmd.randint(1, 100)
    # number validate
    run = True
    i = 0
    while i < 10 and run:
        i += 1
        user = int(input("Guess the Number: "))
        if user < 1 or user > 100:
            print("That is not a valid choice")
        else:
            run = game(user, target)

        if i == 10:
            print(f"the Code was {target}")


main()
```
