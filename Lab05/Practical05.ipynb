# Exercise 1: Data Filtering and Transformation
imagine you have a list of names and their corresponding scores for a class. Filter the list to keep only names of students who scored above 80 and then apply a function to format their names with a congratulatory message.

## Steps:
1. Create a list of name and scores;
2. Define functions for filtering and formatting
3. Apply filter and map to achieve the filtering and transformation
4. Print the result.


```python
# List of names and scores
students = {
    "michael": 80,
    "theo": 70,
    "matt": 50,
    "john": 95,
    "sam": 40
}
```

```python
# Filtering

def filter(students):
    filtered = {}
    for index in students:
        if students[index] >= 80:
            filtered.update({index: students[index]})
    return filtered
```

```python
# formatting the message

def formate(filtered):
    for index in filtered:
        print(f"Congratulations, {index}! You scored {filtered[index]}!")
```

```python
# run time

formate(filter(students))
```

# Exercise 2: Text Processing Pipeline
imagine you need to process a text string without using any additional libraries.
Your task is to implement a text processing pipeline that reads a provided string, converts it to lowercase, removes punctuation, and then splits it into individual words

## Steps:
1. Create a sample String.
2. Define a function to clean and tokenize the text without using external libraries.
3. Apply the function to the text.
4. Print the result.

```python
# my string

sentence = "the quick brown... fox jumps ,,,over the lazy ''''dog"
```

```python
# Clean & tokenize

# i'm aware theirs better methods of doing this
def clean(string):
    for letter in string:
        if letter == "'":
            string = string.replace("'", "")
        elif letter == ".":
            string = string.replace(".", "")
        elif letter == ",":
            string = string.replace(",", "")

    return string
```

```python
# tokenize code

def tokenize(string):
    words = []
    memory = []
    for letter in string:
        if letter == " ":
            words.append("".join(memory).lower())
            memory.clear()
        memory.append(letter)
    return words
```

```python
# process data

print(tokenize(clean(sentence)))
```

# Exercise 3: Game Logic with Higher-Order Function

## Challenge:
Simulate rolling two dice in simple dice game. Use function to define die values (1 to 6) and implement rolling them together to get a total score.

## Steps:
1. Define functions for generating random die values.
2. Define a function to roll both dice and get the total score
3. Use the function to roll and print the result


```python
# import

import random
```

```python
# generate random numbers

def dice():
    return random.randint(1, 6)
```

```python
# add function

def add(a, b):
    return a + b
```

```python
#  game logic?

def game(total):
    if total < 10:
        return f"you lost! with {total}"
    else:
        return f"you won! with {total}"
```

```python
# main

def main():
    for i in range(3):
        print(f"turn {i + 1} : {game(add(dice(), dice()))}")
```

```python
main()
```
