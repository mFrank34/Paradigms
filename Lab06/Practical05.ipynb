* Exercise: Write a Python Function to Find the Sum of Digits Recursively
Read and understand the provided explanation carefully before writing the code

2.2. Complete the function body based on the provided explanation:
* The if statement should check if n is equal to 0.
* If true, return 0 (base case).
* If n is not 0, perform the following steps:
* Extract the last digit using n % 10.
* Make a recursive call to sum_of_digits(n // 10) to find the sum of digits
in the remaining number.
* Return the sum of the last digit and the result of the recursive call (n %
10 + sum_of_digits(n // 10)).

```python
def sum_of_digits(n):
    """
    This function calculates the sum of digits in a non-negative integer n recursively.
    Args:
        n: The non-negative integer for which to calculate the sum of digits.
    Returns:
        The sum of digits in n.
    """
    # Base case
    if n == 0:
        return 0

    # Recursive case
    return (n % 10) + sum_of_digits(n // 10)
```

```python
# example
digits = 15
print(f"The sum of digits of {digits} is {sum_of_digits(digits)}")
```

## 3. Exercise: Write a recursive Python program to find the sum of the first n natural numbers.
For example, if you enter Input: 5
Output: 15


```python
def natural(n):
    if n <= 0:
        print("The natural sum cannot be less than zero.")
    # Base case
    if n == 0:
        return 0
    # Recursive case
    return n + natural(n - 1)
```

```python
# example use-case
digits = 5
print(f"The natural of {digits} is {natural(digits)}")
```
