## Car Parking Price Calculator

```python
def parking_fee(hours):
    if hours <= 0:
        print('Invalid input. Hours must be greater than 0.')
        return
    total = 0
    # Convert hours to minutes
    total_minutes = int(hours * 60)

    # Calculate cost based on minutes
    if total_minutes > 0:
        total += 2  # First 60 minutes (1 hour)
        total_minutes -= 60

    if total_minutes > 0:
        if total_minutes <= 180:  # Next 3 hours (180 minutes)
            total += (total_minutes * 1.5) / 60  # Charge per minute
        else:
            total += (180 * 1.5) / 60 + ((total_minutes - 180) * 1) / 60  # After 4 hours

    # Apply cap
    if total > 12:
        total = 12

    return total

print(parking_fee(6.2))  # Test example
```

## Cinema Ticket Pricing

```python
def cinema_price(age, is_student, day, tickets):
    if tickets <= 0:
        print('Invalid number of tickets')
        return
    price = 10
    if age < 12:
        price = 6
    elif age >= 60:
        price = 7
    if day.lower() == 'tue':
        price -= 2  # Tuesday discount
    if is_student and age >= 12:
        price *= 0.9  # Student discount
    total = price * tickets
    if tickets >= 5:
        total *= 0.95  # Group discount
    return round(total, 2)

print(cinema_price(20, True, 'Tue', 1))  # Test example
```

# BMI Calculator

```python
def bmi_calculator(weight, height):
    if height <= 0 or weight <= 0:
        print('Invalid input. Weight and height must be positive.')
        return
    bmi = weight / (height ** 2)
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 24.9:
        return 'Healthy'
    elif bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'

print(bmi_calculator(70, 1.75))  # Test example
```

## Bus Fare System

```python
# Simplified version of bus fare system
def bus_fare(zones, is_peak):
    if zones not in [1, 2, 3]:
        print('Invalid zone')
        return

    fare = 0

    if zones == 1:
        if is_peak:
            fare = 2.50
        else:
            fare = 2.00

    elif zones == 2:
        if is_peak:
            fare = 3.00
        else:
            fare = 2.40

    elif zones == 3:
        if is_peak:
            fare = 3.50
        else:
            fare = 2.80

    return fare

# Test example
print(bus_fare(2, True))  # Output: 3.0
```
